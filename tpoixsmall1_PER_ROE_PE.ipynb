# P/Eは40以下が望ましい
P/E（株価収益率）は、 株価 ÷ 一株当たり当期純利益という計算式で表せます。株が割安かどうかを判断するための数値になります。どんなに良い企業でも多くの投資家が株を購入してしまうと一株あたりの当期純利益が上がりすぎてしまい、本来の株の価値よりも割高となってしまいます。あくまで指標ですが、P/Eは40以下が好ましいと言われております。しかし、株はネットや新聞の評判などの影響などによりP/Eが100以上になっても上がり続けることがあります。もし、P/Eが40を超えた場合には、いつでも売れる準備をするのが良いかもしれません。P/Eが40を超えたとしても株価が上昇しているのであれば、利を伸ばし、すぐに売らずに持ち続けるのも良いかと思います。

```python
df = pd.read_csv('./csv/TOPIX_small1/498_dataframe20201125.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.rename(columns={'Unnamed: 1':'stocks'})
df = df.groupby(['Date','stocks']).mean()
df
```

```python
value_df = df[(df.per < 10) & (df.roe > 0.1)]       # 割安でクオリティが高い銘柄を抽出

plt.hist(value_df["rt"],bins=100)  
plt.grid()# ヒストグラムの描画
plt.show()

balance = value_df.groupby(level=0).mean().cumsum() # 累積リターンを作成

plt.clf()
plt.grid()
plt.plot(balance["rt"])                             # バランスカーブの描画
plt.show()
```
